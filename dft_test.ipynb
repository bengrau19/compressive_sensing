```python
# Import appropriate packages
import numpy as np
from numpy.fft import fft
from matplotlib import pyplot as plt
```

```python
# Set desired time variable
tmin, tmax, dt = 0, 1, .01

# Optional frequency settings
#f = 1 / dt
#f = 1/(tmax - tmin)
for f in range(1, 100):
    
    # Create time vector
    t = np.arange(tmin, tmax-dt, dt) # start, stop, step
    # Take sine, for the freq, of time vector
    ts = np.sin(2*np.pi*f*t) # + np.random.uniform(0, 2*np.pi))
    # Take dft of the sine wave for the given freq
    fs = fft(ts)
    # Plot each. There should be (and are, verified) two peaks, except at multiples of N/2
    plt.plot(np.abs(fs))
    plt.title(f)
    # Plot to max value of fourier series
    # plt.ylim([0, np.max(np.abs(fs))])
    # Show plot each time (not all on same plot)
    plt.show()

#fmax = 1/dt
#fmin = 1 / (tmax -tmin)
# df should give same # of points as time series
```
